```python
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter
plt.style.use('fivethirtyeight')
```

```python
na_vals = [ 'NA', 'Missing' ]
df = pd.read_csv('Zadowolenie z komunikacji miejskiej mieszkańców Krakowa (Responses).csv', na_values=na_vals)
```

```python
df.drop(columns=['Timestamp', 'Email Address'], inplace=True)
```

```python
def create_pie(name_file):
    fstdf = pd.DataFrame(df[name_file].value_counts('normal'))
    age = list(fstdf.index)

    count = []
    for i in range(fstdf[name_file].count()):
        count.append(fstdf[name_file][i])
    for i in range(len(count)): count[i] *= 100

    inp = input('Maybe you want add a title for legend table?: ')
    if inp == 'Yes' or inp == 'yes':
        name_table = input('Enter a title: ')
        plt.pie(count,
            autopct = '%1.1f%%', pctdistance=1.15)

        plt.legend(age, loc='center right', title=name_table, 
                   bbox_to_anchor=(1, 0, 0.5, 0), fontsize='12')

        plt.title(name_file)
    else:
        plt.pie(count,
            autopct = '%1.1f%%', pctdistance=1.15)

        plt.legend(age, loc='center right', bbox_to_anchor=(1, 0, 0.5, 0), fontsize='12')

        plt.title(name_file)
```

```python
svnthdf = pd.DataFrame(df['Ile czasu dziennie Pan/Pani spędza podróżując transportem? '].value_counts())
time = list(svnthdf.index)

svnth = []
for i in range(svnthdf['Ile czasu dziennie Pan/Pani spędza podróżując transportem? '].count()): 
    svnth.append(svnthdf['Ile czasu dziennie Pan/Pani spędza podróżując transportem? '][i])
for i in range(len(svnth)): svnth[i] *= 100

plt.pie(svnth, autopct = '%1.1f%%',  pctdistance=1.15)

plt.legend(time,loc='center right', bbox_to_anchor=(1, 0, 0.5, 0), fontsize='15')

plt.title('Ile czasu dziennie Pan/Pani spędza podróżując transportem? ')
```

```python
df['Do jakiej dzielnicy Pan/Pani jeździ najczęściej?'] = df['Do jakiej dzielnicy Pan/Pani jeździ najczęściej?'].str.replace('Berlin', 'Stare Miasto')
create_pie('Do jakiej dzielnicy Pan/Pani jeździ najczęściej?')
```

```python
df['Dokąd Pan/Pani jeździ transportem?'] = df['Dokąd Pan/Pani jeździ transportem?'].str.replace('Na Berlin', 'Na spacer') 
df['Dokąd Pan/Pani jeździ transportem?'] = df['Dokąd Pan/Pani jeździ transportem?'].str.replace('Blyadki', 'Na spacer')
df['Dokąd Pan/Pani jeździ transportem?'] = df['Dokąd Pan/Pani jeździ transportem?'].str.replace('spacerować', 'Na spacer')
place_counter = Counter()

for response in df['Dokąd Pan/Pani jeździ transportem?']:
    place_counter.update(response.split(','))
places = []
popularity = []

for item in place_counter.most_common(15):
    places.append(item[0])
    popularity.append(item[1])
places, popularity

plt.barh(places, popularity)
plt.gca().invert_yaxis()


```
