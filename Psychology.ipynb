```python
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter
plt.style.use('fivethirtyeight')
```

```python
na_vals = [ 'NA', 'Missing' ]
df = pd.read_csv('Zadowolenie z komunikacji miejskiej mieszkańców Krakowa (Responses).csv', na_values=na_vals)
```

```python
df.drop(columns=['Timestamp', 'Email Address'], inplace=True)
```

```python
df['Ile Pan/Pani ma lat?'].value_counts('normal')

fstdf = pd.DataFrame(df['Ile Pan/Pani ma lat?'].value_counts('normal'))
age = list(fstdf.index)

count = []
for i in range(fstdf['Ile Pan/Pani ma lat?'].count()):
    count.append(fstdf['Ile Pan/Pani ma lat?'][i])
for i in range(len(count)): count[i] *= 100

plt.pie(count, explode=[0.05, 0, 0, 0],
        autopct = '%1.1f%%', pctdistance=1.15)

plt.legend(age, title='Wiek respondemtów',loc='center right', bbox_to_anchor=(1, 0, 0.5, 0), fontsize='15')

plt.title('Ile Pan/Pani ma lat?')
```

```python
scnddf = pd.DataFrame(df['Kim Pan/Pani jest?'].value_counts('normal'))
sex = list(scnddf.index)

scnd = []
for i in range(scnddf['Kim Pan/Pani jest?'].count()):
     scnd.append(scnddf['Kim Pan/Pani jest?'][i])
for i in range(len(scnd)): scnd[i] *= 100

plt.pie(scnd, explode=[0.05, 0, 0],
        autopct = '%1.1f%%', pctdistance=1.15)

plt.legend(sex, title='Płec respondemtów',loc='center right', bbox_to_anchor=(1, 0, 0.5, 0), fontsize='15')

plt.title('Kim Pan/Pani jest?')
```

```python
eghthdf = pd.DataFrame(df['Do jakiej dzielnicy Pan/Pani jeździ najczęściej?'].value_counts())
eghthdf.drop('Berlin', inplace=True) 
district = list(eghthdf.index)

eghth = []
for i in range(eghthdf['Do jakiej dzielnicy Pan/Pani jeździ najczęściej?'].count()): 
    eghth.append(eghthdf['Do jakiej dzielnicy Pan/Pani jeździ najczęściej?'][i])
for i in range(len(eghth)): eghth[i] *= 100

plt.pie(eghth, autopct = '%1.1f%%',  pctdistance=1.15)

plt.legend(district,loc='center right', bbox_to_anchor=(1, 0, 0.5, 0), fontsize='15')

plt.title('Do jakiej dzielnicy Pan/Pani jeździ najczęściej?')
```

```python
df['Dokąd Pan/Pani jeździ transportem?'] = df['Dokąd Pan/Pani jeździ transportem?'].str.replace('Na Berlin', 'Na spacer') 
df['Dokąd Pan/Pani jeździ transportem?'] = df['Dokąd Pan/Pani jeździ transportem?'].str.replace('Blyadki', 'Na spacer')
df['Dokąd Pan/Pani jeździ transportem?'] = df['Dokąd Pan/Pani jeździ transportem?'].str.replace('spacerować', 'Na spacer')
place_counter = Counter()

for response in df['Dokąd Pan/Pani jeździ transportem?']:
    place_counter.update(response.split(','))
places = []
popularity = []

for item in place_counter.most_common(15):
    places.append(item[0])
    popularity.append(item[1])
places, popularity

plt.barh(places, popularity)
plt.gca().invert_yaxis()


```
